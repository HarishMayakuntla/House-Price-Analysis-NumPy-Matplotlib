```python

house_numbers=np.arange(0,len(price+1))
house_numbers
```

```python
## line graph

plt.figure(figsize=(10,8))
plt.plot(house_numbers,price,marker='s',markersize=6,mfc='red',color="blue",mec='red')
plt.title("House Prices",fontsize=16, fontweight="bold")
plt.xlabel("House Number")
plt.ylabel("Price (₹)")
plt.grid(axis='y')
ax = plt.gca()
ax.set_facecolor('lightyellow')
ax.spines['top'].set_linewidth(3)
ax.spines['bottom'].set_linewidth(3)
ax.spines['left'].set_linewidth(3)
ax.spines['right'].set_linewidth(3)
plt.ticklabel_format(style='plain', axis='y')
plt.savefig("line_chart.png", dpi=300, bbox_inches="tight")
plt.show()
from google.colab import files
files.download("line_chart.png")
```

```python
## bar graph

plt.figure(figsize=(10,8))
plt.bar(house_numbers,
        area,
        color="skyblue",
        edgecolor="black",
        linewidth=1.5,label='house area')
plt.title("House Area Comparison", fontsize=16, fontweight="bold")
plt.xlabel("House Number")
plt.ylabel("Area (sq.ft)")
plt.legend(loc="upper left")
ax = plt.gca()
ax.set_facecolor("lightyellow")
for spine in ax.spines.values():
    spine.set_linewidth(2)
plt.savefig("bar_chart.png", dpi=300)

plt.show()

files.download("bar_chart.png")
```

```python
## scatter

plt.figure(figsize=(10,8))
plt.scatter(
    area,price ,color='red' ,marker='*'
)
plt.title("Area vs Price", fontsize=16, fontweight="bold")
plt.xlabel("Area (sq.ft)")
plt.ylabel("Price (₹)")
ax=plt.gca()
ax.set_facecolor('lightyellow')
ax.spines['top'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
plt.grid(axis='y')
plt.ticklabel_format(style='plain',axis='y')
plt.savefig("scatter_plot.png", dpi=300)
plt.show()
from google.colab import files
files.download("scatter_plot.png")
```

```python
## histogram

plt.figure(figsize=(10,8))
plt.hist(price,bins=8,color="green",
         edgecolor="black")
plt.title("price distrubution ",fontsize=16,fontweight='bold')
plt.xlabel('price(₹)')
plt.ylabel("Number of Houses")
plt.ticklabel_format(style="plain",axis="x")
plt.savefig("histogram_plot.png",dpi=300)
plt.show()
files.download("histogram_plot.png")
```
